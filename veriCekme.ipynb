```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
base_url = "https://www.trendyol.com/cep-telefonu-x-c103498?pi="
```

```python
total_pages = 25
```

```python
fieldnames = ['name', 'price', 'link']
```

```python
with open("products.csv", "w", encoding="utf-8", newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()


    for page in range(1, total_pages + 1):
        url = base_url + str(page) 
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        
        product_cards = soup.find_all('div', class_='p-card-chldrn-cntnr card-border')

        for card in product_cards:
            try:
               
                name = card.find('span', class_='prdct-desc-cntnr-name')
                product_name = name.text.strip() if name else 'No Name'

               
                price = card.find('div', class_='prc-box-dscntd')
                product_price = price.text.strip() if price else 'No Price'

               
                link = card.find('a', href=True)
                product_link = "https://www.trendyol.com" + link['href'] if link else 'No Link'

               
                writer.writerow({
                    'name': product_name,
                    'price': product_price,
                    'link': product_link
                })

                print(f"Processed: {product_name}")

            except Exception as e:
                print(f"Error processing product: {e}")

print("Telefon verileri 'products.csv' dosyasına kaydedildi.")
```
